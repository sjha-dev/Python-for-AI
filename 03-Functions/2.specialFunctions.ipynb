## lambda functions:  Lambda functions are small anonymous functions, meaning they do not have a defined name. These are small, short-lived functions used to pass simple logic to another function.

### 1.Contain only one expression.
### 2.Result of that expression is returned automatically (no return keyword needed).

```python
cret = lambda x:x+1
print(cret(4))
```

```python
crety = lambda x: x*x
print(crety(3))
```

```python
## Lambda functions with multiple arguments ...

crety2 = lambda a ,b,c : a+b+c 
print(crety2(2,3,5))
```

```python
### if else condition with lamda functions 
### wap to check weather the number is positive or negative or zero uing lambda functions 

check = lambda x : "x is positive" if x>0 else "x is negative "
print(check(-78))

```

```python
name='sudhanshu shekhar jha '
upper= lambda a:a.upper()
print(upper(name))
```

### Map function : map() function in Python applies a given function to each element of an iterable (list, tuple, set, etc.) and returns a map object (iterator).
### it is a higher order function.

```python
s=['1','2','3','4','5']
result = map(int ,s)                              ### int function is applied into the s list (each element )
print(result)                                     ### only result gives the map object , if we need a list values then define the data type 
```

```python
s=['1','2','3','4','5']
result = map(int ,s)
print(list(result))
```
