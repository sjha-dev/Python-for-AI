```python
## write a program to check weather a number is even or odd.

def check_even_odd(num):
    if num%2==0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")

print(check_even_odd(13))
            
```

### Types of Function Arguments
### Python supports various types of arguments that can be passed at the time of the function call. Below are types of function argument types:.

```python
##default argument

### Default Arguments: A default argument is a parameter that assumes a default value if a value is not provided in the function call for that argument.

def duffy(a,b=78):
    return a,b

print(duffy(23))
```

```python
##Keyword Arguments

### values are passed by explicitly specifying the parameter names, so the order doesn’t matter.

def nummy(fname,lname):
    print(fname , lname) 

nummy(fname='sudhanshu',lname='shekhar')
nummy(lname='shekhar',fname='sudhanshu')
```

```python
##Positional arguments
### values are assigned to parameters based on their order in the function call.

def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Olivia", 27)

print("\nCase-2:")
nameAge(27, "Olivia")
```

## Nested functions (function inside another function )

```python
def heeo():
    def seeo():
        print("hello sudhanshu")

    seeo()
heeo()        
```

```python
## non-local keyword 
## the nonlocal keyword allows the inner function to modify a variable from the outer function instead of creating a new local copy.

def fun1(): 
    a = 45
    def fun2(): 
        nonlocal a 
        a=54
        print(a)
    fun2()
    print(a)
fun1()

 
```
