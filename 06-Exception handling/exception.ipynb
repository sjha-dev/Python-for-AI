### understanding exception handling !!
#### exception handling in python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program. this lesson will cover the basics of exceptions , including how to use try ,except ,else and finally block.


### WHAT IS AN EXCEPTION ? 
#### Exceptions are events that disrupts the flow of program .they occur when an error is encountered during program execution. common exceptions include

#### 1. ZeroDivisionError : Dividing by zero
#### 2. FileNotFoundError : file not found 
#### 3. ValueERROR : Invalid value
#### 4. TypeError  : Invalid type
#### 5. NameError : not defined 

```python
a=b
```

```python
## when ever any error encountered in the flow of program and we may write custom msg for that error then we use try and except block..
try:
    a=b
except:
    print("the value is not assigned ")    
```

```python
try:
    a=b
except NameError as ex:                     ## Here NameError is the class of that type error so it displayed the default msg. and 'as' is alias.
    print(ex)  
```

```python
## E.g 02 

result=10/0
```

```python
try:
    result=10/0
except ZeroDivisionError as Zd:                                     ## zeroDivisionError is also the class of that type error , displayed it's default msg.
    print(Zd)    
```

```python
try:
    result=10/0
except:                        
    print("any number cannot be divided by zero") 
    print("please enter the denominator other than zero")   
```

```python
try:
    result=1/2
    a=b                                               ### NameError is have been there.
except ZeroDivisionError as zD:                             
    print(zD)    
```

```python
try:
    result=1/2
    a=b                                               ### NameError is have been there.
except ZeroDivisionError as zD:                             
    print(zD)    
except Exception as ex:                                 ### Exception is the parent class of all exceptions (NameError,ZeroDivisionError ,etc ) whenever we called exception then we don't need to specipy the type of error it takes automatically the type of error and displayed the msg. 
    print(ex)  
     
```

```python
## try , except and exception 

try:
    num=int(input("enter the number: "))
    result = 24/num
except TypeError:
    print("please enter the valid type ")
except ZeroDivisionError:
    print("please enter the denominator value more than 0 ")
except Exception as ex:
    print(ex)    
        
```

```python
## try ,except ,else 

try:
    num=int(input("enter the number: "))
    result = 24/num
except TypeError:
    print("please enter the valid type ")
except ZeroDivisionError:
    print("please enter the denominator value more than 0 ")
except Exception as ex:
    print(ex) 
else:                                               ## else block is executed when the try block doesn't contain any type of error , if it contains any type of error then it must show the exception msg.
    print(f"the result is {result}")
```

```python
## try ,except ,else and finally

try:
    num=int(input("enter the number: "))
    result = 24/num
except TypeError:
    print("please enter the valid type ")
except ZeroDivisionError:
    print("please enter the denominator value more than 0 ")
except Exception as ex:
    print(ex) 
else:
    print(f"the result is {result}")
finally:                                ## finally block code always implements whether it is exception or not exception.
    print("execution completed")           
```

```python
## file handling with exception handling

try:
    file=open('example.txt','r')
    content=file.read()
    print(content)
except FileNotFoundError:
    print("there are no such file are there ")    
```
