# Strings in Python

## What is a String?

A string is a sequence of characters enclosed within single quotes (`' '`), double quotes (`" "`), or triple quotes (`''' '''` or `""" """`).

Strings are used to store and manipulate textual data such as names, messages, addresses, and more.

### Examples

```python
name = "Sudhanshu"
city = 'Delhi'
message = """Welcome to Python"""
```

---

## Features of Strings

### 1. Ordered

Strings maintain the order of characters.

```python
text = "Python"
print(text[0])  # P
print(text[1])  # y
```

### 2. Immutable

Strings cannot be changed after they are created.

```python
text = "Python"

# This will generate an error
# text[0] = "J"
```

### 3. Supports Indexing

Each character has an index position.

```python
text = "Python"

print(text[0])   # P
print(text[-1])  # n
```

### 4. Supports Slicing

A part of the string can be extracted using slicing.

```python
text = "Python Programming"

print(text[0:6])  # Python
```

### 5. Allows Duplicate Characters

Duplicate characters are allowed in strings.

```python
text = "programming"
```

### 6. Can Store Any Character

Strings can contain letters, numbers, spaces, and special symbols.

```python
text = "Python 3.13 @ AI"
```

### 7. Iterable

Strings can be traversed using loops.

```python
text = "Python"

for char in text:
    print(char)
```

### 8. Dynamic Length

Strings can have any number of characters.

```python
short = "Hi"
long = "Welcome to Python Programming"
```


````python
## let's start programming ..

name = 'sudhanshu'
print(name)
````

````python
## INDEXING IN STRING 

print(name[2])
print(name[3])
print(name[4])
print(name[5])
````

````python
## iteration in string

for i in range(len(name)):
    print(name[i])
````

````python
## Slicing in python 

## [start:stop:interval]
## stop is not included.


print(name[0:9])
print(name[::2])
````

````python
text = 'the indian army'
print(text.title())
````

````python
## find method in string return the index of find argument 

language = "the python programming"
print(language.find("y"))
````

````python
## Count method returns the numbers or frequency of actual argument paased in count method in given string 
print(language.count("o"))
````

````python
## string membership { in and not in }

language1='the programming in python is easy '
language2='the programming in java is little bit complex'

print("python" in language1)
print("python" in language2)
print("java"not in language1)
print("java" in language1)
````
