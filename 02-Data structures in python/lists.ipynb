```python
## Accessing list elements 

fruits_names=["apple","banana","cherry","mango","kiwi"]
print(fruits_names[0])
```

```python
print(fruits_names[-1])
```

```python
print(fruits_names[0:])
```

```python
print(fruits_names[0:4])
```

```python
## Modifying the list elements
fruits_names
```

```python
## Modifying the list elements

fruits_names[2]="guavava"
fruits_names
```

```python
#list methods 

#Append method  --> adds an extra item to list

fruits_names.append("orange")
fruits_names
```

```python
# Insert method --> adds an extra item at a index of e.g (1,"watermelon")
fruits_names.insert(1,"watermelon")
fruits_names
```
