# 📦 Tuples in Python

## What is a Tuple?

A **Tuple** is an **ordered, immutable collection** of items in Python.

- Uses parentheses `()`
- Ordered
- Immutable (cannot be changed)
- Allows duplicate values
- Can store different data types

### Syntax

```python
tuple_name = (item1, item2, item3)
```

### Example

```python
fruits = ("Apple", "Banana", "Mango")
print(fruits)
```

---

## Accessing Elements

### Positive Indexing

```python
print(fruits[0])
```

### Negative Indexing

```python
print(fruits[-1])
```

### Slicing

```python
numbers = (10, 20, 30, 40, 50)
print(numbers[1:4])
```

---

## Immutability

Tuple elements cannot be modified.

```python
fruits[0] = "Orange"
```

Output:

```
TypeError
```

---

## Tuple Packing

```python
student = "Sudhanshu", 21, "IT"
```

---

## Tuple Unpacking

```python
student = ("Sudhanshu", 21, "IT")

name, age, branch = student
```

---

## Single Element Tuple

```python
t = (5,)
```

Without the comma, it becomes an integer.

---

## Tuple Methods

### count()

```python
numbers = (1,2,2,3,2)

numbers.count(2)
```

### index()

```python
numbers = (10,20,30,40)

numbers.index(30)
```

---

## Tuple vs List

| List | Tuple |
|------|--------|
| Mutable | Immutable |
| Uses `[]` | Uses `()` |
| Slower | Faster |

---

## When to Use Tuple?

Use tuples when data should **not change**, such as:

- Date of Birth
- Coordinates
- RGB Colors
- Employee IDs

Example:

```python
location = (28.6139, 77.2090)
```

---

## Summary

- Ordered collection
- Immutable
- Uses `()`
- Allows duplicates
- Supports indexing and slicing
- Only two methods: `count()` and `index()`

````python
## let's starting tuples 

student = ("sudhanshu","ramesh","sumit","amit","priyanka","reema","sonali","farhan")
print(student)
````

````python
print(type(student))
````

````python
## can store different data types 

coordinates = ("north-west",34.567,56)
print(coordinates)
````

````python
## can store duplicate values 

fruits = ("Mango","guvava","banana","grapes","banana")
print(fruits)
````

````python
## indexing of tuple {positive indexing } / Accessing of element
print(fruits[2])
````

````python
## indexing of tuple {negative indexing } / Accessing of element

print(fruits[-2])
````

````python
## Slicing of elements in the tuple 

print(student[0:6])      ## simple 
print(student[0:6:2])       ## [start : end : interval]
print(student[-4:-1])      ## negative slicing
print(student[::2])       ##[start : end  : interval ]
````

````python
## Iteration on tuple 

for st in student:
    print(st)
````

````python
for i in range(len(student)):
    print(i,"-->",student[i])
````

````python
## Few methods of tuples 

## 1. index

print(student.index("farhan"))

## 2. count 

print(fruits.count("banana"))

## Note : there are only two built-in methods for tuples and others are python buit-in methods which is useful in tuple like len , max , min etc..
````
