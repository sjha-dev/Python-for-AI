### file handling : File handling refers to the process of performing operations on a file, such as creating, opening, reading, writing and closing it through a programming interface. It involves managing the data flow between the program and the file system on the storage device, ensuring that data is handled safely and efficiently.
### Need for File Handling :-
#### Store data permanently, even after the program ends.
#### Access external files like .txt, .csv, .json, etc.
#### Process large files efficiently without using much memory.
#### Automate tasks like reading configs or saving outputs.

```python
### Read a whole file

with open ('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
### read a file line by line 

with open ('example.txt','r') as file:
    for line in file:
        print(line)
        
```

```python
### read a file line by line 

with open ('example.txt','r') as file:
    for line in file:
        print(line.strip())                    ##  strip() --> removes the new line character 
```

```python
## writing a file (overwriting)
with open('example.txt','w') as file:
    file.write("hello guys , this is the overridden text \n")          ## \n --> new line character
    file.write("my name is sudhanshu shekhar jha \n ")

    
```

```python
## writing a file (without overwriting)
with open('example.txt','a') as file:
    file.write("\n this is the appended line ")
    file.write("\n this is the second appended line without overridding")
    
```

```python
## writting a list of lines in a file 

lines=['\n first line \n','second line \n','third line \n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## Binary files

data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
## reading a binary file

with open('example.bin','rb') as file:
    content=file.read()
    print(content)
    
```

```python
## reading the content from source file and write it to a destinanation text file

with open('example.txt','r') as file:
    content = file.read()

with open('destination.txt','w') as file:                       ## if i manually not create the destianation.txt file then with the 'w' mode it create that file and write content accordingly
    file.write(content)    
```

```python
## read a text file and count the number of lines , words and characters.

def count_text_files(file_path):
    with open(file_path,'r') as file:
        lines =file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return (line_count , word_count , char_count)    
        



file_path = 'example.txt'
lines , words , characters = count_text_files(file_path)
print(f'lines: {lines}, words:{words} ,characters:{characters} ')
```

```python
## writing a file first and then read

with open('example.txt','w+') as file:
    file.write("hello jii , how are you \n")
    file.write("what are going on ")

    ## move the file cursor to the begining to read the entire file 
    file.seek(0)

    ## now , read the file 
    content = file.read()
    print(content)
```
