```python
## checking a path is file or directory

path = "package"
if os.path.isfile(path):
    print(f"this '{path}' is a file")
elif os.path.isdir(path):
    print(f"this '{path}' is a directory")
  
else:
    print(f"this '{path}' is neither a file nor a directory ")

```
