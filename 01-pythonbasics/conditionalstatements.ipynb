Conditinoal statements 

```python
## If statement 

age=19

if age>=18:
    print("You are allowed to vote in election")
```

```python
## Else statement
## the else statement executes a block of code when the if condition is false 

age=17
if age>=18:
    print("you are allowed to vote in election ")

else:
    print("You are a minor !")
```

```python
## Elif statement
## The Elif statement allows you to check multiple conditions

age=123

if age<13:
    print("you are a child")

elif age<18:
    print("You are a teenager")

elif age>100:
    print("You are very old person ! happy living ")

elif age>50:
    print("you are a man ! ")   
else:
    print("you are an adult")
```

```python
## Nested conditional statements 
## you can place if , else, elif conditions inside another if , else , elif conditions that's is nested conditional statements !!

## Problem -- you have to check whether the value is even or odd or a negative value  

num = int(input("Enter the number: "))

if num>=0:
    print("the number is positive")

    if num%2==0:
        print(num," is the even number")
    else:
        print(num," is the odd number") 
else:
    print("you have entered a negative value")           
```

```python
## Use cases of conditional statements 

year=int(input("Enter an year : "))

if year>0:
    if year%4==0:
        if year%100==0:
            if year%400==0:
                print(year,"is a leap year")
            else:
                print(year,"is not a leap year")
        else:
            print(year,"is a leap year")
    else:
        print(year,"is not a leap year")      
else:
    print(year,"is not a valid year")                    
```
