```python
range(5)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,4):
    print(i)                                         ## i is the iterator which iterates into the given range (1,4) last number is never included
```

```python
for i in range(1,8,2):
    print(i)                                    ## range(1,8,2) == range(start,stop(not included),step )
```

```python
for i in range(10,1,-2):
    print(i)    
```

```python
## Strings iterates

str="Sudhanshu shekhar jha"
for i in str:
    print(i) 
```

# loops control statements
## break , continue

```python
## While loop
## the while loop continues to execute as long as condition is true.

count=1
while count<=7:
    print(count)
    count=count+1
```

```python
# loop control statements 

## break
## continue
```

```python
number=2
while number%2==0:
    print(number)
    number=number+1
```

```python
#break statement 
## The break loops exits the loop prematurely 

for i in range(5):
    if i==5:
        break
    print(i)
```

```python
# break with print statement

for i in range(7):
    if i==5:
        print("koi baat nhi")
        break
    print(i)
```

```python
#continue statement -- skips the value 

for i in range(7):
    if i==5:
        continue
    print(i)
```

```python
#continue statement with print statement -- for warning or any types of indication 

for i in range(7):
    if i==5:
        print(i,"hota hai laadle !!")
        continue
    print(i)
```

```python
## problem -- write all the odd numbers between 1 to 10 using for loop and continue statement

for i in range(10):
    if i%2==0:
        continue
    print(i)
    
```

```python
## pass --the pass statement is a null operation , it does not thing 

for i in range(10):
    if i==4:
        pass
    print(i) 
```

# Nested loops 
## A loop inside another loop

```python
## Nested for loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
#EXAMPLE -- The sum of first n natural numbers using while loop

n=int(input("Enter the number :"))
sum=0
count=1
while count<=n:
    sum=count+sum
    count=count+1
print(sum)    
```

```python
n=int(input("Enter the number"))
sum=0
for i in range(n+1):
    sum=sum+i
print(sum)    
```
